1. ì–´ë–¤ ë†ìž¥ì—ì„œ ìƒì‚°ëœ ê³„ëž€ 35 ê°œì˜ í‘œë³¸ì„
ë½‘ì•„ì„œ ì•„ëž˜ ë³€ìˆ˜ wì— ì €ìž¥í•˜ì˜€ë‹¤. ê³„ëž€ì˜ í‰ê·  ë¬´ê²Œì— ëŒ€í•œ 95% ì‹ ë¢° êµ¬ê°„ì„

```python
w=[22,
 24,
 23,
 22,
 21,
 27,
 26,
 27,
 25,
 29,
 21,
 25,
 28,
 26,
 20,
 21,
 25,
 20,
 28,
 23,
 26,
 20,
 21,
 27,
 24,
 28,
 25,
 26,
 20,
 24]
```

```python
# ì—¬ê¸°ì— ì½”ë“œë¥¼ ìž‘ì„±í•˜ì‹œì˜¤.
import numpy as np
import scipy.stats as stats

xbar = np.mean(w)
s = np.std(w, ddof=1)
alpha = 0.05
zalpha = stats.norm.ppf(1-alpha/2)
v = zalpha * (s / np.sqrt(len(w)))
min = xbar - v
max = xbar + v
print('min : ', min, 'max : ', max)
```

2. ì–´ë–¤ ì •ì±…ì— ëŒ€í•œ ì°¬ì„±ìœ¨ì„ ì¡°ì‚¬í•˜ê¸° ìœ„í•´ 783ëª…ì„
ëžœë¤í•˜ê²Œ ì„ íƒí•˜ì—¬ ì°¬ì„± ì—¬ë¶€ë¥¼ ì¡°ì‚¬í•˜ì˜€ëŠ”ë°, ì´ ì¤‘ 475 ëª…ì´ ì°¬ì„±í•˜ì˜€ë‹¤. ì°¬ì„±ìœ¨ ð‘ì˜ 95% ì‹ ë¢°êµ¬ê°„ (confidence interval) ì„ êµ¬í•˜ì‹œì˜¤.

```python
# 2.ë²ˆ ë¬¸ì œì— ëŒ€í•œ ì½”ë“œ
x = 475
n = 783
phat = x/n
alpha = 0.05
zalpha = stats.norm.ppf(1-alpha/2)
sd = np.sqrt((phat * (1-phat))/n)
ci = [phat - zalpha*sd, phat + zalpha*sd]
print(ci)
```

```python
# ì½”ë“œ ìž‘ì„±

# H0:mu > 23.5 --> Z > zalpha

# ê°€ì„¤ì— ì˜í•œ ëª¨í‰ê· 
mu = 23.5
# í‘œë³¸í‰ê· 
xbar = np.mean(w)
# í‘œë³¸í‘œì¤€íŽ¸ì°¨
sd = np.std(w, ddof=1)
z = (xbar-mu)/(sd/np.sqrt(len(w)))
print('ê²€ì •í†µê³„ëŸ‰ : ', z)

# ìœ ì˜ìˆ˜ì¤€
alpha = 0.05
# ìž„ê³„ê°’
cri = stats.norm.ppf(1-alpha)
print('ìž„ê³„ê°’ : ', cri)

print(z > cri)
```

```python
import statistics
import scipy
import numpy
import random
random.seed(2021)
a = [random.randrange(10, 1000) for i in range(100)]
```

아래에서 제시하는 내용을 출력하는 코드를 작성하세요. 

* mean, 평균

* median, 중간값

* population variance, 모 분산

* sample variance, 표본 분산

* population standard deviation, 모 표준편차

* sample standard deviation, 표본 표준편차

* range, 범위

* Q1, 제 1사분위수

* Q3, 제 3사분위수

* IQR, 사분위수범위

```python
#여기에 코드를 추가하세요
print("평균 = ", statistics.mean(a))

a.sort()
print(f'중간값 = {statistics.median(a)}' )

print(f'모분산 = {round(statistics.pvariance(a), 4)} in statistics module')
print(f'모분산 = {round(numpy.var(a), 4)} in numpy modoule')

print(f'표본분산 = {round(statistics.variance(a), 4)} in statistics module')
print(f'표본분산 = {round(numpy.var(a, ddof=1), 4)} in numpy modoule')

print(f'모 표준편차 = {round(statistics.pstdev(a), 4)}')
print(f'표본 표준편차 = {round(statistics.stdev(a), 4)}')

print(f'범위 = {max(a) - min(a)}')
print(f'범위 = {numpy.max(a) - numpy.min(a)} in numpy module')

print(f'Q1, 제 1사분위수 = {numpy.quantile(a, .25)}')
print(f'Q3, 제 3사분위수 = {numpy.quantile(a, .75)}')
print(f'IQR, 사분위범위 = {numpy.quantile(a, .75) - numpy.quantile(a, .25)}')  # Q3 - Q1
```

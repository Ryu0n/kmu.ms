2. í‰ê· ì´ 20ì´ê³ , í‘œì¤€í¸ì°¨ê°€ 5ì¸ ì •ê·œ ë¶„í¬ì—ì„œ ë‹¤ìŒ ê° í™•ë¥ ì„ êµ¬í•˜ëŠ” python ì½”ë“œë¥¼ ì‘ì„±í•˜ì‹œì˜¤.

```python
m, s = 20, 5

# ğ‘ƒ[ğ‘‹â‰¥30]
x_30 = stats.norm.cdf(30, loc=m, scale=s)
p = 1 - x_30
print(p)

# ğ‘ƒ[ğ‘‹<15]
x_15 = stats.norm.cdf(15, loc=m, scale=s)
p = x_15
print(p)

# ğ‘ƒ[10<ğ‘‹<25]
x_25 = stats.norm.cdf(25, loc=m, scale=s)
x_10 = stats.norm.cdf(10, loc=m, scale=s)
p = x_25 - x_10
print(p)
```
